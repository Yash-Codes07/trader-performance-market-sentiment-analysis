#### 2. Conversion of dates and alignment of both datasets

```python
fear_df["date"] = pd.to_datetime(fear_df["date"]).dt.normalize()
fear_df["date"].head()
```

```python
trade_df["Timestamp IST"] = pd.to_datetime(trade_df["Timestamp IST"], dayfirst=True)
trade_df["date"] = trade_df["Timestamp IST"].dt.normalize()
trade_df.head()
```

```python
merged_df = pd.merge(trade_df, fear_df[["value", "classification", "date"]] , on="date", how="left")
merged_df = merged_df.rename(columns={"value" : "Value", "classification" : "Classification", "date" : "Date"})
merged_df.head()
```

```python
merged_df = merged_df.dropna(subset=["Classification"])
```

```python
merged_df['Classification'].value_counts()
```

#### 3. Key Metrics

##### (i) Daily PnL Per Trader/User

```python
trader_PnL_df = merged_df.groupby(["Account", "Date"])["Closed PnL"].sum().reset_index()
trader_PnL_df = trader_PnL_df.rename(columns={"Closed PnL" : "Daily PnL"})
trader_PnL_df.head()
```

##### (ii) Win Rate, Average Trade Size

```python
merged_df["Win"] = merged_df["Closed PnL"] > 0
```

```python
win_rate_df = merged_df.groupby("Account").agg(Winning_Trade=("Win", "sum"), Total_Trade=("Win", "count")).reset_index()
win_rate_df["Win Rate (%)"] = (win_rate_df["Winning_Trade"] / win_rate_df["Total_Trade"]) * 100
win_rate_df.head()
```

```python
avg_trade_df = merged_df.groupby("Account")["Size USD"].mean().reset_index()
avg_trade_df = avg_trade_df.rename(columns={"Size USD" : "Avg Trade Size (USD)"})
avg_trade_df.head()
```

##### (iii) Leverage Distribution

##### Note
The provided historical trader dataset does not contain a leverage field. Therefore, leverage distribution could not be analyzed directly. The remaining requested metrics were computed using the available data.

##### (iv) Number Of Trades Per Day

```python
trades_per_day_df = merged_df.groupby("Date")["Trade ID"].count().reset_index()
trades_per_day_df = trades_per_day_df.rename(columns={"Trade ID" : "Number of Trades"})
trades_per_day_df.head()
```

##### (v) Long/Short Ratio

```python
long_trades = (merged_df["Direction"].isin(("Open Long", "Close Long"))).sum()
short_trades = (merged_df["Direction"].isin(("Open Short", "Close Short"))).sum()
print("Number of long trades :",long_trades)
print("Number of short trades :",short_trades)

long_short_ratio = long_trades/short_trades
print("Long/Short ratio :",long_short_ratio)
```

```python
long_short_df = pd.DataFrame({
    "Long Trades": [long_trades],
    "Short Trades": [short_trades],
    "Long/Short Ratio": [long_short_ratio]
})
long_short_df
```
